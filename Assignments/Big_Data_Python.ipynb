```python
# Cleaning Data
# Remove all rows with the missing values

new_edu = edu.dropna(how='any', subset=['Value'])
new_edu.head()
```

```python
edufilled = edu.fillna(value = {'Value': 0})
edufilled
```

```python
edufilledmean = edu.fillna(value = {'Value': edu['Value'].mean()})
edufilledmean.tail(10)
```

```python
edufilledmode = edu.fillna(value = {'Value': edu['Value'].mode()})
edufilledmode.tail()
```

```python
# Sort Data
edu.sort_values(by='Value', ascending=True, inplace=True)
edu.head()
```

```python
edu.sort_values(by='Value', ascending=False, inplace=True)
edu.head()
```

```python
edu.sort_index(axis = 0, inplace = True)
edu.head()
```

```python
group = edu[['GEO', 'Value']].groupby('GEO').mean()
group.tail()
```

```python
group = edu[['GEO', 'Value']].groupby('GEO').max()
group.tail()
```
