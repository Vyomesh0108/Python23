```python
# Combine both data frames into a single one: Columns in the second data frame to the RIGHT of the first data frame

print("Combined both data frames into a single one", end="\n")

combined_dataframe = pd.concat([csv_data, json_data], axis=1)
combined_dataframe
```

```python
# Export your results as a CSV and make sure it reads back into Python properly.

combined_dataframe.to_csv('combined_data.csv', index=False)
```

```python
# Reading the newly created dataframe

print("Exported the result as a CSV and reading back the newly created CSV file", end="\n")
reading_new_df = pd.read_csv('combined_data.csv')
reading_new_df
```

```python
# Showing the list of columns in your dataframe

print("Showing the list of columns in the dataframe", end="\n")
reading_new_df.columns
```

```python
# Do you find any duplicated columns? Display the first 5 rows of your data, is your data accurate?

print("Duplicate Columns", end="\n\n")
duplicate_columns = reading_new_df.columns[reading_new_df.columns.duplicated()]
print("Duplicate Columns: ",duplicate_columns, end="\n\n")

print("First 5 Rows of the data", end="\n\n")
reading_new_df.head()
```

```python
# Use merge from the pandas package to make an inner join.

print("Doing inner join by using merge", end="\n\n")
merged_dataframe = pd.merge(csv_data, json_data, on='species_id', how='inner')
merged_dataframe
```

```python
# Fill the missing value for the column weigh by the mean

print("Filling missing value for the weigh column by mean")
merged_dataframe['weight'] = merged_dataframe['weight'].fillna(merged_dataframe['weight'].mean())
merged_dataframe
```

```python
# Plot the distribution of taxa by sex by plot

import matplotlib.pyplot as plt
import seaborn as sb

print("Plot the Distribution of taxa by sex by plot", end="\n\n")
sb.countplot(x='taxa', hue='sex', data=merged_dataframe)
plt.xlabel('Taxa')
plt.ylabel('Count')
plt.title('Distribution of Taxa by Sex')
plt.show()
```

```python
# Plot the distribution of taxa by plot

import matplotlib.pyplot as plt
import seaborn as sb

print("Plot Distribution of taxa by plot", end="\n\n")
plt.figure(figsize=(8,6))
plt.hist(merged_dataframe['taxa'], bins=20)
plt.xlabel('Taxa')
plt.ylabel('Count')
plt.title('Distribution of Taxa')
plt.show()
```
