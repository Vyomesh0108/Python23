```python
# Rename the columns belonging to Products by removing Mnt from them.
# a. You should have: 'Wines','Fruits','Meat','Fish','Sweet','Gold'

print("Renamed the columns belonging to Products by removing Mnt from them", end="\n\n")
customer_data.rename(columns = {'MntWines': 'Wines', 'MntFruits': 'Fruits',
                               'MntMeatProducts': 'MeatProducts', 'MntFishProducts': 'FishProducts',
                               'MntSweetProducts': 'SweetProducts', 'MntGoldProds': 'GoldProds'}, inplace=True)

customer_data.columns
```

```python
# Rename the columns belonging to Place by removing Num and Purchases from them.
# a. You should have: 'Web', 'Catalog', 'Store'

print("Renamed the columns belonging to place by removing Num and Purchases from them", end="\n\n")
customer_data.rename(columns = {'NumWebPurchases': 'Web', 'NumCatalogPurchases': 'Catalog', 
                                'NumStorePurchases': 'Store'}, inplace=True)

customer_data.columns
```

```python
# Keep only NumDealsPurchases column in the Promotion section

print("Kept the NumDealsPurchases column in the Promotion section", end="\n")
promotion_section = customer_data[['NumDealsPurchases']]
promotion_section
```

```python
# Delete the column NumWebVisitsMonth

print("Deleted the NumWebVisitsMonth column", end="\n")
customer_data = customer_data.drop(columns = ['NumWebVisitsMonth'])
customer_data.columns
```

```python
# The attribute Year_Birth is not helpful to segment your data. Instead create a new derived column named Age

print("Created new derived column named Age", end="\n")
customer_data['Age'] = 2023 - customer_data['Year_Birth']
customer_data
```

```python
# Create a new column named Spending that holds the total amount spent on all products categories.

print("Created a new column name Spending which holds the total amount spent on all products", end="\n")
customer_data['Spending'] = (customer_data['Wines'] + customer_data['Fruits'] + customer_data['MeatProducts'] + 
                            customer_data['FishProducts'] + customer_data['SweetProducts'] + 
                            customer_data['GoldProds'])

customer_data
```

```python
# Change the Dt_Customer type from object to datetime. Use the format YYYY-MM-DD.

print("Changed the Dt_Customer type from object to datetime", end="\n")
customer_data['Dt_Customer'] = pd.to_datetime(customer_data['Dt_Customer'], format='%Y-%m-%d')
customer_data.dtypes
```

```python
# Show the unique values of the column data Marital_Status.

print("Showing the unique values of the Martial_Status data", end="\n")
print(customer_data['Marital_Status'].unique())
```

```python
# It is better to work with lower categorical values. Hence, weâ€™ll classify customers in two segments for
# their martial status: change the column Marital status as follows:
# a. Change the values: Divorced, Single, Absurd and Widow to 'Alone'
# b. Change Married and Together to 'Not Alone'

print("Changed the values of Martial status column", end="\n")
customer_data['Marital_Status'] = customer_data['Marital_Status'].replace(['Divorced', 'Single', 'Absurd',
                                                                          'Widow', 'YOLO'], 'Alone')
customer_data['Marital_Status'] = customer_data['Marital_Status'].replace(['Married', 'Together'], 'Not Alone')
customer_data['Marital_Status']
```

```python
# Show the unique values of the column Education

print("Showing the unique values of the Education column", end="\n")
print(customer_data['Education'].unique())
```

```python
# We want to segment the customers in 2 groups based on their education level: change the column Education as follows:
# a. Change the values: 'Basic' and '2n Cycle' to 'Undergraduate',
# b. All other values to 'Postgraduate'

print("Changed the values of Education column", end="\n")
customer_data['Education'] = customer_data['Education'].replace({'Basic': 'Undergraduate', 
                                                                 '2n Cycle': 'Undergraduate'})
customer_data['Education'] = customer_data['Education'].replace({'PhD': 'Postgraduate', 'Master': 'Postgraduate',
                                                                'Graduation': 'Postgraduate'})
customer_data['Education']
```

```python
# Have a look at column Income. Investigate the presence of outliers? Delete them accordingly.

import matplotlib.pyplot as plt
import seaborn as sb

sb.boxplot(customer_data['Income'])
plt.show()

print("Looked at the Income column", end="\n")
customer_data = customer_data[customer_data['Income'] < 100000]
print(customer_data['Income'] < 100000)
```

```python
# Create a new column named Children to hold the total number of children for every customer.

print("Created new column Children which has total number of children", end="\n")
customer_data['Children'] = customer_data['Kidhome'] + customer_data['Teenhome']
customer_data
```

```python
# Create a new column Has_Complaint as follows:
# a. 'Has complaint' if the customer complained in the last 2 years,
# b. Otherwise 'No complaint'

import numpy as np

print("Created new column Has_Complaint", end="\n")
customer_data['Has_Complaint'] = np.where(customer_data['Complain'] == 1, 'Has complaint', 'No complaint')
customer_data
```

```python
# Use our cleaned data frame (in part II) to create a new one with the columns
import matplotlib.pyplot as plt
import seaborn as sb

new_dataframe = customer_data[['Age','Spending','Marital_Status','Education','Income',
                               'Children','Dt_Customer','Has_Complaint']]
print(new_dataframe)

# Distribution of age
sb.set_style('whitegrid')
plt.figure(figsize=(14,8))
plt.title("Age")
sb.displot(new_dataframe['Age'])
plt.show()


# Distribution of spending
sb.set_style('whitegrid')
plt.figure(figsize=(14,8))
plt.title("Spending")
sb.displot(new_dataframe['Spending'])
plt.show()

# Distribution of Customer Marital Status
sb.set_style('whitegrid')
plt.figure(figsize=(14,8))
plt.title("Marital Status")
sb.countplot(x='Marital_Status', data=new_dataframe, palette='RdBu_r')
plt.show()

# Distribution of Customer Education Level
sb.set_style('whitegrid')
plt.figure(figsize=(14,8))
plt.title("Education")
sb.countplot(x='Education', data=new_dataframe, palette='inferno')
plt.show()

# Distribution of Customer Income Level
sb.set_style('whitegrid')
plt.title("Income")
plt.figure(figsize=(14,8))
sb.displot(new_dataframe['Income'])
plt.show()


# Distribution of Customer with Children or Not
sb.set_style('whitegrid')
plt.figure(figsize=(14,8))
plt.title("Children")
sb.countplot(x='Children', data=new_dataframe, palette='inferno')
plt.show()

# Distribution of Customer Complaints
sb.set_style('whitegrid')
plt.figure(figsize=(14,8))
plt.title("Complaints")
sb.countplot(x='Has_Complaint', data=new_dataframe, palette='inferno')
plt.show()
```
